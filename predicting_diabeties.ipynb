# Abstract

Diabetes is a major health concern that impacts around 38.4 million Americans as of 2021, and close to 8.7 million cases went undiagnosed due to limited and insufficient healthcare access. However, the traditional screening methods like A1C tests and glucose tolerance tests were once very expensive and needed some labor activities or medical devices to use, which made these methods more difficult and inaccessible for many people. To solve this issue, we decided to focus our project on developing a machine learning model that could predict and inform us whether someone is non-diabetic, pre-diabetic, or diabetic based on simple and easy to measure health indicators. Based on publicly available health data, we were curious about and also focused on determining if factors like BMI, high blood pressure, and cholesterol levels could be good predictors of diabetes. 

To do this, we first used the Diabetes Health Indicators Dataset from The Behavioral Risk Factor Surveillance System (BRFSS) 2015 dataset and cleaned the data through checking for missing data and doing transformations on variables. After that, we tried to study the characteristics of non-diabetics, pre-diabetics, and diabetics through visualizations and dataframe aggregations between the diabetics classification column and the rest of our feature columns. Lastly, we used a Random Forest Classifier to classify whether an individual was a non-diabetic, pre-diabetic, or diabetic based on the most important features, using Grid Search to do hyperparameter tuning.

After working with the dataset, our results demonstrate that health indicators such as BMI, age, high blood pressure, and high cholesterol were the most significant predictors of diabetes. This result aligns with and supports our hypotheses and findings. One interesting finding from our project was that factors like heart disease, stroke, and physical activity also played an important role in predicting diabetes, which we didn't expect this earlier. In terms of the overall performance of our model, the model we built achieved an overall accuracy of 75%, but it had trouble accurately classifying pre-diabetic and diabetic cases. Based on our analysis, this is likely because the dataset was missing key features like family history and dietary information. These limitations show that while machine learning has potential for diabetes risk assessment, adding more detailed health data could help improve the model’s performance.

# Hypothesis


We hypothesize that Body Mass Index (BMI) will be the strongest predictor of Type 2 diabetes risk. A higher BMI, which is closely linked to poor dietary habits (high in fats, carbs, and sugars), leads to glucose spikes and increased insulin demand, contributing to a higher risk of developing diabetes. Age is also a strong metric because frequently older people have a greater chance of developing diabetes.

We also expect high cholesterol and high blood pressure to be strongly positively correlated to diabetes as they are often associated with metabolic syndrome and insulin resistance, which drive the progression from normal glucose metabolism to prediabetes and type 2 diabetes.

Overall, because of the strongly correlated features we have with diabetes, we believe it is possible to build an accurate machine learning model that predicts the category of type 2 diabetes: non-diabetic, pre-diabetic, and diabetic.

Because we have 3 categories, let's do an ANOVA test to see if there's a significant different. We fulfilled assumptions of ANOVA: the 3 distributions are roughly normally distributed as we can see above.

Null Hypothesis: There is no significant difference in the mean BMI values between the three diabetes categories. That is any observed differences in means are due to random chance.

Alternative Hypothesis: At least one of the groups has a significantly different mean BMI compared to the others.

```python

group1 = diabetes_df[diabetes_df['Diabetes_012'] == 0]['BMI']
group2 = diabetes_df[diabetes_df['Diabetes_012'] == 1]['BMI']
group3 = diabetes_df[diabetes_df['Diabetes_012'] == 2]['BMI']

f_statistic, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

As we can the p-value is highly statistically significant < 0.01. We reject the null. So, there is at least one group that has significant difference in mean than the other groups. Let's do Tukey's honestly significant difference (HSD) test that is often used to determine which groups have a significant difference after an ANOVA test detects a significant result.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey HSD test
tukey_result = pairwise_tukeyhsd(diabetes_df['BMI'], diabetes_df['Diabetes_012'], alpha=0.05)

print(tukey_result)
```

In all 3 cases, we reject the null(there is a difference in means). So, BMI could be useful feature in determining diabeties diagnosis.

Let's see if there is a significant difference between the groups. We know our data is not normally distributed, so we can't use ANOVA. So, we will use the Kruskal-Wallis Test since it has the same function as ANOVA but does not use normality and is non-parametric. Note that the kruskal-wallis compares the distributions but not particularly the means.


Null Hypothesis: There is no significant difference in the distribtuion of mental health values between the three diabetes categories. That is any observed differences in means are due to random chance.

Alternative Hypothesis: At least one of the groups has a significantly different distribution compared to the others.

```python
import scikit_posthocs as sp
group1 = diabetes_df[diabetes_df['Diabetes_012'] == 0]['MentHlth']
group2 = diabetes_df[diabetes_df['Diabetes_012'] == 1]['MentHlth']
group3 = diabetes_df[diabetes_df['Diabetes_012'] == 2]['MentHlth']


H, p_value = stats.kruskal(group1, group2, group3)  
print(f"p-val = {p_value}")
if p_value < 0.05:
    dunn_results = sp.posthoc_dunn([group1, group2, group3], p_adjust='bonferroni')  
    print(dunn_results)
```

We got a p-value of less than 0.01 indicating highly statistically significant results. We reject the null hypothesis. We can then use Dunn's Test to see which pairs are significantly different in which we see all pairs get a p-value of less than 0.01. So, there is a difference.

```python
selected_features = ["HighBP", "HighChol", "HeartDiseaseorAttack", "BMI", "GenHlth","Age","DiffWalk","Stroke", "PhysActivity"]
X = diabetes_df[selected_features]
y = diabetes_df["Diabetes_012"]
```

```python
scaler = StandardScaler() # standardizing our features
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
# spliting our dataset with a test size of 0.2
```

Also, note that we used a method called SMOTE(Synthetic Minority Oversampling Technique) to balance the target classes. This is too make sure our data is not biased to the high majority class, which is non-diabetics. This will adjust the frequency of classes, while preserving the original feature distributions and relationships within the data, helping our model learn meaningful patterns from minority classes.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

Because we are doing a classification model, we want to use some kind of classifier. A very popular and good classifier is a random forest classifiers, which uses ensemble learning. This model tries n decision trees with some way of making each one different through bootsrapping training data and selecting a subset of features for each splitting of the node. Anyway, this is a good model for getting low variance and low bias. Below, we using GridSearchCV, which uses 5-fold validation set, to find the best hyperparameters for the best performance of our model. 

```python
param_grid = {
    'n_estimators': [300],#150,200 tried both 300 best
    'max_depth': [None,10], # more than 10 is worse so kept these 2
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1, 2]
}

rf_model = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf_model, param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train_resampled, np.array(y_train_resampled))
```

```python
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(X_test)
```

Below is the best hyperparameters from the possible options typed above, accuracy, and classification report.

```python
print("Best Parameters:", grid_search.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
